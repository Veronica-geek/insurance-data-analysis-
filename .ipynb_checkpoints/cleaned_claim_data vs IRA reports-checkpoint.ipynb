```python
def extract_ira_data(pdf_path, quarter):
    # Extract all tables using tabula
    tables = tabula.read_pdf(pdf_path, pages="all", multiple_tables=True, lattice=True)
    
    # Combine tables into one DF for appendices (insurer-level data)
    insurer_df = pd.concat(tables, ignore_index=True)
    
    # Clean insurer DF: Standardize columns, remove NaNs, convert numerics
    if not insurer_df.empty:
        insurer_df.columns = insurer_df.columns.str.strip().str.replace('\r', ' ').str.replace('\n', ' ')
        insurer_df = insurer_df.dropna(how='all').reset_index(drop=True)
        numeric_cols = [col for col in insurer_df.columns if 'Claims' in col or '%' in col]
        for col in numeric_cols:
            insurer_df[col] = insurer_df[col].astype(str).str.replace(',', '').str.replace('%', '').str.strip()
            insurer_df[col] = pd.to_numeric(insurer_df[col], errors='coerce')
    
    # Extract industry totals from text (using regex on PDF text if needed; assume from last row of tables)
    # For simplicity, grab the 'Industry' row from the appendices
totals = []
categories = ['Liability', 'Non-Liability', 'Long-Term']
for table in tables:
    if table.empty:
        continue
    table_str = table.to_string().lower()  # Lowercase for case-insensitive matching
    cat = None
    if 'liability' in table_str:
        cat = 'Liability'
    elif 'non-liability' in table_str:
        cat = 'Non-Liability'
    elif 'long-term' in table_str:
        cat = 'Long-Term'
    if cat:
        # Find 'Industry' row (case-insensitive)
        industry_rows = table[table.apply(lambda row: row.astype(str).str.lower().str.contains('industry').any(), axis=1)]
        if not industry_rows.empty:
            total_row = industry_rows.iloc[0].copy() # Take first matching
            total_row['Quarter'] = quarter
            total_row['Category'] = cat
            totals.append(total_row)
totals_df = pd.DataFrame(totals) if totals else pd.DataFrame()
```

```python
import pandas as pd

claims_df['REGISTRATION_DATE'] = pd.to_datetime(claims_df['REGISTRATION_DATE'], errors='coerce')

# Create 'Quarter' column in 'YYYY-Q#' format 
claims_df['Quarter'] = claims_df['REGISTRATION_DATE'].dt.to_period('Q').astype(str)
```

```python
# Status flags for clarity and correctness
claims_df['is_paid'] = claims_df['KC_STATUS_TEXT'].str.contains('Authorized Payment|Ready for Payment|Accept Releasing Letter', na=False)
claims_df['is_declined'] = claims_df['KC_STATUS_TEXT'].str.contains('Repudiated|Initial Rejected|Incomplete', na=False)
claims_df['is_closed_no_claim'] = claims_df['KC_STATUS_TEXT'].str.contains('Claim Closed|Claim Cancelled|Close Claim In Between', na=False)

# Aggregate company claims by quarter/category
# To approximate IRA: Actionable = Outstanding_Begin + Intimated + Revised (per quarter group)
company_agg = claims_df.groupby(['Quarter', 'Category']).agg(
    Actionable_Count=('KC_STATUS_TEXT', 'count'),
    Intimated_Count=('TIME_TAKEN_REGISTRATION', lambda x: (x <= 1).sum()),
    Revised_Count=('KC_FIRST_REPUDIATE_REM', lambda x: x.notna().sum()),
    Paid_Count=('is_paid', 'sum'),
    Declined_Count=('is_declined', 'sum'),
    Closed_No_Claims_Count=('is_closed_no_claim', 'sum'),
    Outstanding_Count=('KC_STATUS_TEXT', lambda x: x.str.contains('INPROCESS|Registered|Reinspection Appointed', na=False).sum()),
    
    
    Paid_Amount=('KC_SUM_INSURED', lambda x: x[claims_df.loc[x.index, 'is_paid']].sum()),
    Declined_Amount=('KC_OUTSTANDING_RESERVE', lambda x: x[claims_df.loc[x.index, 'is_declined']].sum()),
    Closed_No_Claims_Amount=('KC_OUTSTANDING_RESERVE', lambda x: x[claims_df.loc[x.index, 'is_closed_no_claim']].sum()),
    Outstanding_Amount=('KC_OUTSTANDING_RESERVE', 'sum')
).reset_index()
```

```python
# Calculate ratios (matching IRA)
company_agg['Payment_Ratio'] = (company_agg['Paid_Count'] / company_agg['Actionable_Count']) * 100
company_agg['Declined_Ratio'] = (company_agg['Declined_Count'] / company_agg['Actionable_Count']) * 100
company_agg['Closed_Ratio'] = (company_agg['Closed_No_Claims_Count'] / company_agg['Actionable_Count']) * 100
company_agg['Intimated_Ratio'] = (company_agg['Intimated_Count'] / company_agg['Actionable_Count']) * 100
company_agg['Revised_Ratio'] = (company_agg['Revised_Count'] / company_agg['Actionable_Count']) * 100

print(company_agg)  # For verification
```

```python
# Clustering

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Use only available diff columns
available_diffs = [col for col in ['Payment_Ratio_Diff', 'Declined_Ratio_Diff', 'Closed_Ratio_Diff'] if col in comparison_df.columns]

if len(available_diffs) == 0:
    print("No difference columns available for clustering — skipping")
else:
    print(f"Clustering using: {available_diffs}")
    
    cluster_df = comparison_df[available_diffs].fillna(0)
    
    # Standardize
    scaler = StandardScaler()
    scaled = scaler.fit_transform(cluster_df)
    
    # Cluster (3 groups)
    kmeans = KMeans(n_clusters=min(3, len(cluster_df)), random_state=42, n_init=10)
    comparison_df['Performance_Cluster'] = kmeans.fit_predict(scaled)
    
    # Label clusters by average performance
    cluster_means = comparison_df.groupby('Performance_Cluster')[available_diffs].mean().mean(axis=1)
    max_cluster = cluster_means.idxmax()
    min_cluster = cluster_means.idxmin()
    
    comparison_df['Cluster_Label'] = comparison_df['Performance_Cluster'].map(
        lambda x: 'Strong Outperformer' if x == max_cluster
        else ('Consistent Underperformer' if x == min_cluster else 'Average Performer')
    )
    
    # Color palette
    palette = {'Strong Outperformer': 'green', 'Average Performer': 'gold', 'Consistent Underperformer': 'red'}
    
    # Plot (use first two diffs if more than one)
    x_col = available_diffs[0]
    y_col = available_diffs[1] if len(available_diffs) > 1 else available_diffs[0]
    
    plt.figure(figsize=(12, 8))
    sns.scatterplot(
        data=comparison_df,
        x=x_col,
        y=y_col,
        hue='Cluster_Label',
        style='Category',
        s=200,
        palette=palette,
        alpha=0.9
    )
    plt.axhline(0, color='gray', linestyle='--', alpha=0.7)
    plt.axvline(0, color='gray', linestyle='--', alpha=0.7)
    plt.title('Performance Clustering vs IRA Benchmarks\n(Positive = Better than Industry)')
    plt.xlabel(x_col.replace('_Diff', ' Difference'))
    plt.ylabel(y_col.replace('_Diff', ' Difference'))
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
    
    # Summary
    print("\nCluster Summary:")
    print(comparison_df.groupby(['Cluster_Label', 'Category'])['Quarter'].count().unstack(fill_value=0))
```

```python
# Linear Regression Forecast: Payment Ratio & Reserves 

if 'Time_Index' not in comparison_df.columns:
    comparison_df = comparison_df.sort_values(['Quarter', 'Category']).reset_index(drop=True)
    comparison_df['Time_Index'] = np.arange(len(comparison_df))

fig, axes = plt.subplots(1, 2, figsize=(18, 7))

# Liability only
liability_df = comparison_df[comparison_df['Category'] == 'Liability'].copy()

if liability_df.empty:
    print("No Liability data")
else:
    X = liability_df['Time_Index'].values.reshape(-1, 1)
    future_idx = np.arange(len(liability_df), len(liability_df) + 4).reshape(-1, 1)
    future_quarters = ['2026Q1', '2026Q2', '2026Q3', '2026Q4']

    # Payment Ratio
    model_pay = LinearRegression().fit(X, liability_df['Payment_Ratio'])
    pred_pay = model_pay.predict(future_idx)

    axes[0].plot(liability_df['Quarter'], liability_df['Payment_Ratio'], 'o-', color='blue', label='Company', linewidth=4, markersize=10)
    axes[0].plot(liability_df['Quarter'], liability_df['Payment_Ratio_Num'], 'x--', color='orange', label='IRA Benchmark', linewidth=3)
    axes[0].plot(future_quarters, pred_pay, '^:', color='green', label='Forecast', linewidth=4, markersize=12)
    axes[0].set_title('Liability Payment Ratio Forecast', fontsize=16, pad=20)
    axes[0].set_ylabel('Payment Ratio (%)', fontsize=12)
    axes[0].legend(fontsize=12)
    axes[0].grid(True, alpha=0.3)
    axes[0].tick_params(axis='x', rotation=45)

    # Reserves — NOW IN THOUSANDS KES (not millions)
    y_reserves_thousands = liability_df['Outstanding_Amount'] / 1000
    model_res = LinearRegression().fit(X, y_reserves_thousands)
    pred_res_thousands = model_res.predict(future_idx)

    axes[1].plot(liability_df['Quarter'], y_reserves_thousands, 'o-', color='blue', label='Company', linewidth=4, markersize=10)
    if 'Outstanding_End_Amount' in liability_df.columns:
        ira_thousands = liability_df['Outstanding_End_Amount'] / 1000
        axes[1].plot(liability_df['Quarter'], ira_thousands, 'x--', color='orange', label='IRA Industry', linewidth=3)
    axes[1].plot(future_quarters, pred_res_thousands, '^:', color='green', label='Forecast', linewidth=4, markersize=12)
    axes[1].set_title('Liability Outstanding Reserves Forecast\n(Thousands KES)', fontsize=16, pad=20)
    axes[1].set_ylabel('Reserves (Thousands KES)', fontsize=12)
    axes[1].legend(fontsize=12)
    axes[1].grid(True, alpha=0.3)
    axes[1].tick_params(axis='x', rotation=45)

plt.suptitle('LIABILITY CLAIMS: Current Performance & 2026 Forecast vs IRA Industry', fontsize=18, y=1.02)
plt.tight_layout()
plt.show()

# Print readable forecast
print("Liability Forecast (2026):")
print(f"   Payment Ratio: {pred_pay.round(1).tolist()}%")
print(f"   Reserves: {pred_res_thousands.round(0).astype(int).tolist()} Thousand KES")
print(f"   → Trend: {'Increasing' if model_res.coef_[0] > 0 else 'Decreasing'} by {abs(model_res.coef_[0]):.0f}K per quarter")
```
